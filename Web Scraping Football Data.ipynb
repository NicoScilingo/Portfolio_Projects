```python
#Import Libreries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By


# Define URL to scrape
website = 'https://www.adamchoi.co.uk/overs/detailed'

# Define ChromeDriver Path
path = r'C:\Users\nico_\OneDrive\Escritorio\DataScience\Python\ChromeDriver\chromedriver-win64\chromedriver.exe'
service = Service(path)
```

```python
# Start ChromeDriver
driver = webdriver.Chrome(service=service)

# Open the website
driver.get(website)
```

```python
# Finds 'All matches' item
all_matches_button = driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')

# Clicks on All Matches
all_matches_button.click()
```

```python
matches = driver.find_elements(By.TAG_NAME, 'tr')

for match in matches:
    print(match.get_attribute('outerHTML'))  # Verifies what is there in 'match'
```

```python
# We create a list to append the data. 

date = []
home_team = []
away_team = []
score = []

for match in matches:
    try:
        date.append(match.find_element(By.XPATH, './td[1]').text)  
        home_team.append(match.find_element(By.XPATH, './td[2]').text)  
        score.append(match.find_element(By.XPATH, './td[3]').text) 
        away_team.append(match.find_element(By.XPATH, './td[4]').text) 
    except Exception as e:
        print(f"Error al procesar match: {e}") # Noticed there was an error with trs that only contained 1 td
```

```python
for match in matches:
    tds = match.find_elements(By.XPATH, './td')  # gettin all the tds from tr

    # Verifying if rows have exactly 4 <td> and if there are no empty values for the columns
    if len(tds) == 4 and all(td.text.strip() != '' for td in tds):
        date.append(tds[0].text.strip())      
        home_team.append(tds[1].text.strip())
        score.append(tds[2].text.strip())
        away_team.append(tds[3].text.strip())
    else:
        print(f"Ignored row: {match.text} - Number of <td>: {len(tds)}")

# Printing lengh to confirm all have the same length
print(f"Length of date: {len(date)}")
print(f"Length of home_team: {len(home_team)}")
print(f"Length of score: {len(score)}")
print(f"Length of away_team: {len(away_team)}")


```

```python
# importing the DF as an excel
df = pd.DataFrame({
    'date': date,
    'home_team': home_team,
    'score': score,
    'away_team': away_team
})
```

```python
# Making sure 'date' is in datetime format, specifying day goes first
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')  # Convertirting to datetime

# Making sure score is string
df['score'] = df['score'].astype(str)  # Converting to string
```
