```python
#Import Libreries
import pandas as pd
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time



# Define URL to scrape
website = 'https://www.adamchoi.co.uk/overs/detailed'

chrome_options = Options()
chrome_options.add_argument("--headless")  # Activate headless mode

# Define ChromeDriver Path
path = r'C:\Users\nico_\OneDrive\Escritorio\DataScience\Python\ChromeDriver\chromedriver-win64\chromedriver.exe'
service = Service(path)
```

```python
# Start ChromeDriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the website
driver.get(website)
```

```python
# Selecting from Dropdowns
dropdown = Select(driver.find_element(By.ID, 'country'))
time.sleep(1)
dropdown.select_by_visible_text('Argentina')
```

```python
# Before clicking on all matches set to to wait 3 seconds to load all the Argentina info
time.sleep (3)
```

```python
# Finds 'All matches' item
all_matches_button = driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')

# Clicks on All Matches
all_matches_button.click()
```

```python
matches = driver.find_elements(By.TAG_NAME, 'tr')

for match in matches:
    print(match.get_attribute('outerHTML'))  # Verifies what is there in 'match'
```

```python
# We create a list to append the data. 

date = []
home_team = []
away_team = []
score = []

for match in matches:
    try:
        date.append(match.find_element(By.XPATH, './td[1]').text)  
        home_team.append(match.find_element(By.XPATH, './td[2]').text)  
        score.append(match.find_element(By.XPATH, './td[3]').text) 
        away_team.append(match.find_element(By.XPATH, './td[4]').text) 
    except Exception as e:
        print(f"Error al procesar match: {e}") # Noticed there was an error with trs that only contained 1 td
```

```python
date.clear()
home_team.clear()
score.clear()
away_team.clear()

# Printing lengh to confirm all items have been deleted
print(f"Length of date: {len(date)}")
print(f"Length of home_team: {len(home_team)}")
print(f"Length of score: {len(score)}")
print(f"Length of away_team: {len(away_team)}")
```

```python
for match in matches:
    tds = match.find_elements(By.XPATH, './td')  # gettin all the tds from tr

    # Verifying if rows have exactly 4 <td> and if there are no empty values for the columns
    if len(tds) == 4 and all(td.text.strip() != '' for td in tds):
        date.append(tds[0].text.strip())      
        home_team.append(tds[1].text.strip())
        score.append(tds[2].text.strip())
        away_team.append(tds[3].text.strip())
    else:
        print(f"Ignored row: {match.text} - Number of <td>: {len(tds)}")

# Printing lengh to confirm all have the same length
print(f"Length of date: {len(date)}")
print(f"Length of home_team: {len(home_team)}")
print(f"Length of score: {len(score)}")
print(f"Length of away_team: {len(away_team)}")


```

```python
# importing the DF as an excel
df = pd.DataFrame({
    'date': date,
    'home_team': home_team,
    'score': score,
    'away_team': away_team
})
```

```python
# Making sure 'date' is in datetime format, specifying day goes first
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')  # Convertirting to datetime

# Making sure score is string
df['score'] = df['score'].astype(str)  # Converting to string
```

```python
print(df.dtypes)


# Although data types were set correctly, excel was converting the score column into dates when possible.
```

```python
# Solving the above issue

# Saving date column as date
date_column = df['date']

# Creating new df for the string/text columns
text_df = df[['home_team', 'score', 'away_team']].copy()

# Ensuring that the 'score' column is a string
text_df['score'] = text_df['score'].astype(str)

# Normalizing team names
def normalize_string(s):
    return s.replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u')

text_df['home_team'] = text_df['home_team'].apply(normalize_string)
text_df['away_team'] = text_df['away_team'].apply(normalize_string)

# Dividing the 'score' column into two new columns: 'home_goals' and 'away_goals'
text_df[['home_goals', 'away_goals']] = text_df['score'].str.split('-', expand=True)

# Converting columns to numeric type
text_df['home_goals'] = pd.to_numeric(text_df['home_goals'], errors='coerce')
text_df['away_goals'] = pd.to_numeric(text_df['away_goals'], errors='coerce')

# Combining text_df with the 'date' column
result_df = pd.concat([date_column, text_df[['home_team', 'home_goals', 'away_goals', 'away_team']]], axis=1)

# Exporting the final DataFrame, ensuring that the format of the columns is respected
result_df.to_csv(
    r'C:\Users\nico_\OneDrive\Escritorio\DataScience\Python\Projects\Web Scraping with Selenium\Partidos_Futbol_Argentino.csv',
    index=False,
    quoting=csv.QUOTE_NONNUMERIC,
    date_format='%Y-%m-%d'  # This is optional and can be adjusted according to your needs
)
```
